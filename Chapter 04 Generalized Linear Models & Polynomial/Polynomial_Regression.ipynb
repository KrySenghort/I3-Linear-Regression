# <p style='text-align:center'><b><font color = green|>Polynomial Regression Algorithm

<font size=4>Polynomial Regression is a form of regression analysis in which the relationship between the independent variable(s) and the dependent variable is modeled as an nth-degree polynomial. In other words, it fits a polynomial equation to the data points to create a curve that best represents the relationship between the variables.

The general form of a polynomial regression equation is:</font>

### $$ y = b_0 + b_1x + b_2x^2 + b_3x^3 + ... + b_nx^n + ε $$ 

<font size=4><font color = green|>where:
- $ y $ is the dependent variable (the predicted value).
- $ x $ is the independent variable (the input feature).
- $ b_0, b_1, b_2, ...,$ bn are the coefficients of the polynomial terms.
- $ n $ is the degree of the polynomial (n can be any positive integer).
- $ ε $ represents the error term, accounting for the variability not explained by the model.</font>

In polynomial regression, the algorithm tries to find the best-fitting polynomial curve that minimizes the difference between the predicted values and the actual values of the dependent variable.

The proof for polynomial regression lies in the process of finding the coefficients $ b_0, b_1, b_2, ...,b_n $ that minimize the sum of squared errors (SSE). The SSE is the sum of the squares of the differences between the predicted values and the actual values.

Mathematically, the SSE can be represented as follows:</font>


###  $$ SSE = Σ(y_{actual} - y_{predicted})^2 $$


<font size=4>The goal is to find the coefficients that minimize SSE. To do this, we take the derivative of the SSE with respect to each coefficient and set it to zero. By solving these equations, we can find the values of 

$ b_0, b_1, b_2, ...,b_n $ that minimize SSE and provide the best-fitting polynomial curve.

The exact proof for polynomial regression can be quite involved, as it involves calculus and optimization techniques. The process of solving for the coefficients is typically done using methods like the method of least squares, which is used in linear regression as well.

While I can't provide a complete mathematical proof here, the key idea is to use calculus to find the optimal coefficients that minimize the sum of squared errors, resulting in the best polynomial fit to the data points. This process allows us to construct a polynomial regression model and use it for prediction and analysis in scenarios where the relationship between the variables is better represented by a polynomial curve.</font>

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

# Generate some sample data for demonstration
np.random.seed(0)
X = np.sort(5 * np.random.rand(80, 1), axis=0)
y = np.sin(X).ravel() + np.random.randn(80) * 0.1

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, random_state=0)

# Create a polynomial feature transformer with degree 3
degrees = [1, 2, 3, 4, 5, 6, 7]

mse_list = []
deg_list = []

for degree in degrees:
    poly_features = PolynomialFeatures(degree=degree)

    # Transform the input features to polynomial features
    X_train_poly = poly_features.fit_transform(X_train)
    X_test_poly = poly_features.transform(X_test)

    # Create and fit a linear regression model on the polynomial features
    model = LinearRegression()
    model.fit(X_train_poly, y_train)

    # Make predictions on the test data
    y_pred = model.predict(X_test_poly)

    # Calculate the Mean Squared Error (MSE) to evaluate the model
    mse = mean_squared_error(y_test, y_pred)
    print(f"Mean Squared Error: {mse}")

    deg_list.append(degree)
    mse_list.append(mse)

    # Plot the results
    plt.scatter(X, y, s=20, label="Data")
    plt.plot(X_test, y_pred, color='red',
             label=f"Polynomial Regression (Degree {degree})")
    plt.xlabel('X')
    plt.ylabel('y')
    plt.legend()
    plt.show()

df = pd.DataFrame({'degree': deg_list, 'mse': mse_list})
df
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

# Load the Iris dataset
iris = pd.read_csv(
    'https://raw.githubusercontent.com/uiuc-cse/data-fa14/gh-pages/data/iris.csv')

# Use 'sepal length' as the independent variable (X) and 'sepal width' as the dependent variable (y)
X = iris['sepal_length'].values.reshape(-1, 1)
y = iris['sepal_width'].values

# Split the data into training and testing sets (80% for training, 20% for testing)
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, random_state=0)

# Choose the degree of the polynomial you want to use
degrees = np.arange(50)
mse_list = []
deg_list = []

for degree in degrees:
    # Create a polynomial feature transformer with the chosen degree
    poly_features = PolynomialFeatures(degree=degree)

    # Transform the input features to polynomial features
    X_train_poly = poly_features.fit_transform(X_train)
    X_test_poly = poly_features.transform(X_test)

    # Create and fit a linear regression model on the polynomial features
    model = LinearRegression()
    model.fit(X_train_poly, y_train)

    # Make predictions on the test data
    y_pred = model.predict(X_test_poly)

    # Calculate the Mean Squared Error (MSE) to evaluate the model
    mse = mean_squared_error(y_test, y_pred)
    print(f"Mean Squared Error: {mse}")

    deg_list.append(degree)
    mse_list.append(mse)

    # Plot the original data points
    plt.scatter(X, y, s=20, label="Data")

    # Sort X_test for a smoother plot
    X_test_sorted = np.sort(X_test, axis=0)

    # Predict the corresponding y values for the sorted X_test data
    y_pred_sorted = model.predict(poly_features.transform(X_test_sorted))

    # Plot the polynomial regression line
    plt.plot(X_test_sorted, y_pred_sorted, color='red',
             label=f"Polynomial Regression (Degree {degree})")
    plt.xlabel('Sepal Length')
    plt.ylabel('Sepal Width')
    plt.legend()
    plt.show()

df = pd.DataFrame({'degree': deg_list, 'mse': mse_list})
df
```
