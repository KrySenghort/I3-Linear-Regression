```python
# install.packages("HistData")
library(HistData)
data(Galton)
Galton = data.frame(Galton)
View(Galton)
```

```python
library(dplyr)
glimpse(Galton)
```

<font color = blue><b><i>a.) Reconstruct the contingency table between the height of 928 adults children and the average height of their 205 set of parents.</font></b></i>

```python
cont.parent.child <- table(Galton$child, Galton$parent)
library(knitr)
kable(cont.parent.child)
```

<i><b><font color = blue>b. Reconstruct the scatter plot of and regression line between the height of children and average height of their parents.</font></b></i>

```python
plot(Galton$parent, Galton$child, 
     xlab="average of height of parents in inch",
     ylab="height of children")
abline(lm(Galton$child ~ Galton$parent))
```

```python
# Alternative way of plot using ggplot2 packages
# install.packages("ggplot2")
library(ggplot2)
ggplot(Galton, aes(x=parent, y=child)) +
  geom_point() + xlab("average of height of parents in inch") + 
  ylab("height of children") +
  ggtitle("Scatter Plot of Average Height of Parents vs. Height of Children")+ 
  geom_smooth(method = "lm", se = FALSE) +  
  theme_bw()
```

Figure 1: Scatter plot including a regression line between the height of children and the average height of
their parents

```python
# install.packages("gamlss.data") #for the first time you need to install the package
library(gamlss.data)
# loading data into R environment
data(rent99)
rent99<-data.frame(rent99)
# to view data set
View(rent99)
```

```python
# to get the overview of the data set
glimpse(rent99)
```

```python
# more information of data set
help("rent99") # or ? rent 99
par(mfrow=c(2,2))
```

<i><b><font color = blue>a. Reconstruct the histograms and kernel density estimates below.</font></b></i>

```python
library(dplyr)
hist(rent99$rent,breaks=60,freq=FALSE,
     xlab="net rent in Euro",
     ylab="estimated density",
     main="histogram of net rent in Euro")
lines(density(rent99$rent),lwd=2,col="red")
```

```python
hist(rent99$rentsqm,breaks=60,freq=FALSE,
     xlab="net rent per sqm in Euro",
     ylab="estimated density",
     main="Histogram of area in square meter")
lines(density(rent99$rentsqm),lwd=2,col="red")
```

```python
hist(rent99$area, breaks=60, freq = FALSE,
     xlab = "area in sqm", 
     ylab = "estimated density",
     main = "Histogram of Area in Square Meter") 
lines(density(rent99$area), lwd = 2, col = "red")
```

```python
hist(rent99$yearc, breaks=60, freq = FALSE,
     xlab = "year of construction", 
     ylab = "estimated density",
     main = "Histogram of Year of Construction") 
lines(density(rent99$yearc), lwd = 2, col = "red")
```

```python
hist(rent99$rent, breaks=60, freq = FALSE,
     xlab = "net rent in Euro", 
     ylab = "estimated density",
     main = "Histogram of Net Rent in Euro") 
lines(density(rent99$rent), lwd = 2, col = "red")

hist(rent99$rentsqm, breaks=60, freq = FALSE,
     xlab = "net rent per sqm in Euro", 
     ylab = "estimated density",
     main = "Histogram of Net Rent Per Square Meter in Euro") 
lines(density(rent99$rentsqm), lwd = 2, col = "red")

hist(rent99$area, breaks=60, freq = FALSE,
     xlab = "area in sqm", 
     ylab = "estimated density",
     main = "Histogram of Area in Square Meter") 
lines(density(rent99$area), lwd = 2, col = "red")

hist(rent99$yearc, breaks=60, freq = FALSE,
     xlab = "year of construction", 
     ylab = "estimated density",
     main = "Histogram of Year of Construction") 
lines(density(rent99$yearc), lwd = 2, col = "red")
```

Figure 2: Histogram and kernel density estimators for the continuous variables rent, rentsqm, area, and
yearc
